```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 경고 메시지를 무시하도록 설정
import warnings
warnings.simplefilter("ignore")
```

```python
from matplotlib import font_manager, rc

# 사용하려는 OTF 폰트의 경로 지정
font_path = r'C:\Users\User\AppData\Local\Microsoft\Windows\Fonts\NotoSansKR-Regular.otf'

# 폰트 이름
font_name = font_manager.FontProperties(fname=font_path).get_name()

# 폰트 설정
rc('font', family=font_name)
```

```python
# 엑셀 파일을 읽어와서 데이터프레임으로 변환
df = pd.read_excel('../excelNgraph_img/Excel/전자상거래_경험_여부.xlsx')

# 데이터프레임 정보 출력 (열 이름, 데이터 타입, 크기 등 확인)
df.info()

# 데이터프레임의 상위 5개 행을 출력
df.head()
```

```python
# '구분별1' 열에서 '연령'으로 필터링한 데이터프레임 생성
byAge = df.query("구분별1 == '연령'")

# 첫 번째 열 제거 (불필요한 열 제거)
byAge = byAge.drop(byAge.columns[0], axis=1)

# '경험있다'와 '경험없다'의 2년간 평균값을 계산하고 '차이1', '차이2' 열 추가
byAge = byAge.assign(
    경험있다 = lambda x: (x['2021.1'] + x['2022.1']) / 2,  # 2021년과 2022년 '경험있다' 평균
    경험없다 = lambda x: (x['2021.2'] + x['2022.2']) / 2,  # 2021년과 2022년 '경험없다' 평균
    차이1 = lambda x: x['2022.1'] - x['2021.1'],  # 2021년과 2022년 '경험있다' 차이
    차이2 = lambda x: x['2022.2'] - x['2021.2']  # 2021년과 2022년 '경험없다' 차이
)

# 처리된 데이터프레임 출력
byAge
```

```python
# 연령대별 2년간 데이터를 포함한 데이터프레임 생성
byAgeMean = byAge.iloc[:, [0, 2, 3, 5, 6, 7, 8, 9, 10]]

# 컬럼 이름을 명확하게 수정
byAgeMean.columns = ['연령대', '있다 2021년', '없다 2021년', '있다 2022년', '없다 2022년', '경험있다', '경험없다', '2년간 경험한 비율 차이', '2년간 경험하지 않은 비율 차이']

# 처리된 데이터프레임 출력
byAgeMean
```

```python
# 그래프를 2개로 나눠서 생성, 각각의 그래프 크기를 설정
fig, ax = plt.subplots(ncols=2, figsize=(18,5))

# 그래프 사이의 간격 설정
fig.subplots_adjust(wspace=0.7)

# 색상 설정 (연령대별로 다른 색상 사용)
cm = sns.color_palette('hls', len(byAgeMean['연령대'].unique()))

# 막대그래프의 너비 설정
width = 0.6

# 첫 번째 그래프: 연령대별 전자상거래 경험 증가세
grp1 = sns.barplot(data=byAgeMean, x='연령대', y='2년간 경험하지 않은 비율 차이', palette=cm, ax=ax[0])

# 첫 번째 그래프의 타이틀 설정
grp1.set_title('2021~2022 연령대별 전자상거래 경험 증가세', pad=40)

# 첫 번째 그래프에 추가 설명 텍스트 추가
grp1.text(6.52, 7.2, 'y축 값 = 2022년 값 - 2021년 값', va='top', ha='right', fontsize=10)
grp1.text(6.525, 6.6, '감소세 = - (증가세)', va='top', ha='right', fontsize=10)

# x축과 y축 설정 (글자 크기, 축 두께, 축 길이 등)
grp1.tick_params(labelsize=12, width=1.5, length=10)
grp1.tick_params(axis='x', labelrotation=45)

# x축과 y축 라벨 설정
grp1.set_xlabel('연령대', labelpad=10)
grp1.set_yticks(np.arange(-2,8,2))
grp1.set_ylabel('경험있다', labelpad=10)

# 범례 제거
grp1.legend([], [], frameon=False)

# 막대 너비를 조정하여 중앙에 배치
for bar in grp1.patches:
    x = bar.get_x()
    old_width = bar.get_width()
    bar.set_width(width)
    bar.set_x(x + (old_width - width) / 2)

# 두 번째 그래프: 연령대별 전자상거래 경험 여부를 산포도로 표시
grp2 = sns.scatterplot(data=byAgeMean, x='경험있다', y='경험없다', s=180, hue='연령대', palette=cm, ax=ax[1])

# 두 번째 그래프의 타이틀 설정
grp2.set_title('2021~2022 연령대별 평균 전자상거래 경험 여부', pad=30)

# x축과 y축 설정 (글자 크기, 축 두께, 축 길이 등)
grp2.tick_params(labelsize=12, width=1.5, length=10)
grp2.set_xticks(np.arange(10, 101, 10))
grp2.set_xlabel('경험있다', labelpad=10)
grp2.set_yticks(np.arange(0, 91, 15))
grp2.set_ylabel('경험없다', labelpad=15)

# 범례 위치와 크기 설정
grp2.legend(loc='center left', bbox_to_anchor=(-0.6, 0.5), fontsize=14)

# 그래프 출력
plt.show()
```
